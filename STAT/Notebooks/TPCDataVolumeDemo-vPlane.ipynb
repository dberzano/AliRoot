## General consideration 
* __*To run this notebeook environment variable AliExternalInfoCache has to point to cache directory*__
* __*In case AliExternalInfoCachedirectory  not defined - data will be cached locally*__

* Qurying data from web server is slow - local chache is used instead
  * Env. variable AliExternalInfoCache used as a cache directory
  * In case variable AliExternalInfoCache not defined - data cached in current working directory
* Chain functionality not always working - in case granularity of trees change
  * __different branchs -> crash of the chaing creation__
  * problem to be solved
     * adding empty branches
     * rerunning creation of materialized views
  * problem with some type of indexed arrays - differnt beahviour in ROOT5 and ROOT6
  
* In order to use chains of trees,  trees have to be cached before
  * To cache data sources for this tutorial (do not forget to set AliExternalInfoCache variable):
  * This can take few minutes. For tutorials AliExternalInfo cache directory could be used
  ```
  AliExternalInfo info;
  info.CacheProduction(TPRegexp("LHC15.*"),TPRegexp(""),"QA.TPC;QA.TRD;QA.ITS;QA.EVS;Logbook; QA.rawTPC");
  ```
  * Working with one period (tree), caching is done automatically
* To initialize metdata - metadata compiled macros are  used. In case macroses not properly compiled, metadata not defined.
  * in case metadata not defined axis description and memebr queriyn not working

## 2. Initialize AliExternalInfo DB 
For this tutorial following sources are needed (For detailed information - see AliExternalInfo tutorial)
* Output can be verbose - in case some source does not work properly we can increase verbosity varaible e.g to 7 
* QA.TPC and QA.rawTPC  source
    * to acces TPC QA and TPC calibration  parameters
* QA.EVS 
    * in this example - to access interaction rate
* Logbook
    * in this example to access TPC data volume
    
* Define derived varaibles combining information
  * normalizedData volume $$normDataVolume=\frac{Logbook.size(MBy)}{QA.EVS.nEvents}$$

````c++
//period="LHC18*"; pass="cpass1_pass1"; ///- enable this line to analyze all 2018
info.fVerbose=7;
treeTPC = info.GetTree("QA.TPC",period,pass,"QA.rawTPC;QA.TPC;QA.EVS;Logbook;Logbook.detector:TPC:detector==\"TPC\"");
//treeTPC = info.GetChain("QA.TPC",period,pass,"QA.rawTPC;QA.TPC;QA.EVS;Logbook;Logbook.detector");
treeTPC->SetMarkerStyle(21);
treeTPC->SetAlias("normDataVolume","Logbook.detector_TPC.bytesInjectedPhysics/Logbook.detector_TPC.eventCountPhysics/1000000");
````

## Query list of the variables of given class
* __TFormula based queries on metadata attributes__
  * Usefull for web interface for filtering the list of variale and aliases for drawing querying
* first list friend trees  
* query TPC dEdx variables
* query TPC DCAr information - which is not of type 
* Logbook queries of variables
  *  using (class) metadata
  *  using TTree::Show or TTree::Print

````c++
treeTPC->GetListOfFriends()->ls();
````

````c++
//  Simple query TPC dEdx variables
AliTreePlayer::selectMetadata(treeTPC->GetFriend("QA.TPC"), "[class==\"TPC&&dEdx\"]",0)->Print();
//  Simple query TPC efficiency variables
AliTreePlayer::selectMetadata(treeTPC->GetFriend("QA.TPC"), "[class==\"TPC&&Eff\"]",0)->Print();
````

````c++
//  Refined query TPC DCAz variables which are of class Err
AliTreePlayer::selectMetadata(treeTPC->GetFriend("QA.TPC"), "[class==\"TPC&&(DCAz&&!Err)\"]",0)->Print();
//  Refined query TPC DCAz variables which are not of class Err
AliTreePlayer::selectMetadata(treeTPC->GetFriend("QA.TPC"), "[class==\"TPC&&(DCAz&&Err)\"]",0)->Print();
````

````c++
// Select logboook time information source
AliTreePlayer::selectMetadata(treeTPC->GetFriend("Logbook"), "[class==\"Logbook&&Time\"]",0)->Print();  
//  Select logboook event stat  information source
AliTreePlayer::selectMetadata(treeTPC->GetFriend("Logbook"), "[class==\"Logbook&&Stat\"]",0)->Print();
````

````c++
// Simple tree functionality to dump tree content and show list of variables (and data)
treeTPC->GetFriend("Logbook.detector_TPC")->Show(1);
treeTPC->GetFriend("Logbook")->Show(0);
````

### 3. Fit dependece of data voume as function of Interaction rate
1.  Define variable normalized data volume using Loogbook 
2.  Make a histogram of data volume as function of iteraction rate
3.  Fit linear model  normalized data volume as function of rate $$Dv=p_{0}+p_{1}*I$$
  1.  Non robust fit using the TProfile
  2.  Creating graphs using  TStatToolkit::MakeGraphErrors and fitting using robust LTM fits

````c++
/// A.) Non robust fit using the TProfile
treeTPC->SetAlias("normDataVolume","Logbook.detector_TPC.bytesInjectedPhysics/Logbook.detector_TPC.eventCountPhysics/1000000");
entries = treeTPC->Draw("normDataVolume:QA.EVS.interactionRate","","prof");
//grNorm=TStatToolkit::MakeGraphErorrs(treeTPC,"normDataVolume:QA.EVS.interactionRate","1",25,1);
treeTPC->GetHistogram()->Fit("pol1","");
canvasDraw->Draw();
````

````c++
/// Creating graphs using TStatToolkit::MakeGraphErrors and fitting using robust LTM fits
gStyle->SetOptFit(1);
grNorm=TStatToolkit::MakeGraphErrors(treeTPC,"normDataVolume:QA.EVS.interactionRate","1",25,1);
grNorm->Draw("ap");
grNorm->Fit("pol1","rob=0.8","");
treeTPC->SetAlias("normDataVolumeFit",TString::Format("(%f+%f*QA.EVS.interactionRate)",pol1->GetParameter(0), pol1->GetParameter(1)).Data());
canvasDraw->Draw();
````
